```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV file
heart_data = pd.read_csv("C:/Users/mrshw/OneDrive/Desktop/heart_attack_prediction_dataset.csv")
heart_data.head()
```

```python
# List the columns
heart_data.columns
```

```python
# Create a dataframe with columns to total certain categories.
heart_totals_data = heart_data[["Age", "Sex", "Obesity", "Continent", "Heart Attack Risk"]]
heart_totals_data.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(heart_totals_data))
tick_locations = [value for value in x_axis]
```

```python
pdf.pivot(index="Continent",columns="Heart Attack Risk",values="count").plot(kind="bar",stacked=True)
```

```python
# Get total settlements by union
continent_data = heart_totals_data["Continent"].value_counts()
```

```python
# Configure plot, figsize, title, and axis labels
figure1 = continent_data.plot(kind="bar", facecolor="red", figsize=(10,8),
                                title="Male/Female Participants by Continent",
                                xlabel="Continent",
                                ylabel="Sex")

# Configure x-tick rotation
xticklabels = continent_data.index
figure1.set_xticklabels(xticklabels, rotation=45, rotation_mode="anchor", ha="right", wrap=True)


# Show plot
plt.show()
```
