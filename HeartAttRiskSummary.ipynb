```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from pathlib import Path
```

```python
# Read in Heart Attack Risk data set
data = Path("C:/Users/mrshw/OneDrive/Desktop/heart_attack_prediction_dataset.csv")
df = pd.read_csv(data)
df.head()
```

```python
# Get the information on the dataframe
df.info()
```

```python
# Determine which measure of central tendency is most appropriate to describe the Ages of those in the dataset
plt.hist(df['BMI'])
plt.xlabel('BMI')
plt.ylabel('Counts')
plt.show()
print(df['BMI'].mean())
print(df['BMI'].median())
print(df['BMI'].mode())
```

```python
# Determine if the age in dataset is normally distributed using a small and large sample size. 
plt.hist(df['Age'])
plt.xlabel('Age (years)')
plt.ylabel('Counts')
plt.show()
print(st.normaltest(df["Age"].sample(100)))
print(st.normaltest(df["Age"].sample(2000)))
```

```python
# Determine if there are any potential outliers in the average income in the dataset
quartiles = df['Cholesterol'].quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of Cholesterol is: {lowerq}")
print(f"The upper quartile of Cholesterol is: {upperq}")
print(f"The interquartile range of Cholesterol is: {iqr}")
print(f"The the median of Cholesterol is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")

outlier_Cholesterol = df.loc[(df['Cholesterol'] < lower_bound) | (df['Cholesterol'] > upper_bound)]
outlier_Cholesterol
```
